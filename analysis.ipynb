```python
print(dff['Class'].value_counts())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=dff, x='Class')
plt.title("Class Distribution (0 = Legit, 1 = Fraud)")
plt.show()
```

```python
fraud_ratio = dff['Class'].mean()
print(f"Fraudulent transactions: {fraud_ratio*100:.4f}%")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(dff['Amount'], bins=100, ax=ax[0], kde=True)
ax[0].set_title("Transaction Amount Distribution")
sns.histplot(dff['Time'], bins=100, ax=ax[1], kde=True)
ax[1].set_title("Transaction Time Distribution")
plt.show()
```

```python
corr = dff.corr()
plt.figure(figsize=(20,15))
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Correlation Matrix")
plt.show()
```

```python
corr_target = corr["Class"].sort_values(ascending=False)
print("Top correlated features with Class:\n", corr_target.head(10))
```
